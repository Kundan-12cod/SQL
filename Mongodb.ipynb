```python
#01  Python script to load Superstore dataset into MongoDB:

import csv
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['superstore']
collection = db['orders']

with open('superstore.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        collection.insert_one(row)
```

```python
#02 Retrieve and print all documents from the Orders collection:




for document in collection.find():
    print(document)
```

```python
#03 Count and display the total number of documents in the Orders collection:



print(collection.count_documents({}))
```

```python
#04 Query to fetch all orders from the "West" region:




for order in collection.find({"Region": "West"}):
    print(order)
```

```python

#05 Query to find orders where Sales is greater than 500:



for order in collection.find({"Sales": {"$gt": 500}}):
    print(order)
```

```python
#06 Fetch the top 3 orders with the highest Profit:



for order in collection.find().sort("Profit", -1).limit(3):
    print(order)
```

```python
#07 Update all orders with Ship Mode as "First Class" to "Premium Class":



collection.update_many({"Ship Mode": "First Class"}, {"$set": {"Ship Mode": "Premium Class"}})
```

```python
#08 Delete all orders where Sales is less than 50:




collection.delete_many({"Sales": {"$lt": 50}})
```

```python
#09 Aggregation to group orders by Region and calculate total sales per region:




pipeline = [
    {"$group": {"_id": "$Region", "total_sales": {"$sum": "$Sales"}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```

```python
#10 Fetch all distinct values for Ship Mode:



ship_modes = collection.distinct("Ship Mode")
print(ship_modes)
```

```python
#11.Count the number of orders for each category:



pipeline = [
    {"$group": {"_id": "$Category", "count": {"$sum": 1}}}
]
for result in collection.aggregate(pipeline):
    print(result)
```
